## Data Cleaning & EDA

###Relationship between room types and price:

```python
# check data type
print(airbnb_df[['room_type', 'price']].dtypes)
```

```python
# convert price to float
airbnb_df['price'] = airbnb_df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt

# Filter out FutureWarnings
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

plt.figure(figsize=(10, 7))
sns.barplot(data=airbnb_df, x="room_type", y="price", ci=None)

plt.title("Relationship between Room Types and Price")
plt.xlabel("Room Types")
plt.ylabel("Average Price")
plt.show()
```

###Relationship between beds and price:

```python
plt.figure(figsize=(10, 7))
sns.barplot(data=airbnb_df, x="beds", y="price", ci=None)
plt.title("Relationship between Beds and Price")
plt.xlabel("Number of Beds")
plt.ylabel("Price")
plt.show()
```

### Relationship between room types, beds number, and price

```python
# Create a catplot to visualize the relationship
g = sns.catplot(data=airbnb_df, x="room_type", y="price", hue="beds", kind="bar", ci=None, height=7, aspect=2)

# Set plot title and labels
g.fig.suptitle("Relationship between Room Types, Beds Number, and Price", y=1.02)
g.set_axis_labels("Room Types", "Average Price")
plt.show()

```
